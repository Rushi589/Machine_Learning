```python
df["arrival_year"]=pd.to_datetime(df["arrival_date"]).dt.year
```

```python
df["arrival_month"]=pd.to_datetime(df["arrival_date"]).dt.month
```

```python
df["arrival_date"]=pd.to_datetime(df["arrival_date"]).dt.day
```

```python
df.head()
```

```python
df.drop("arrival_date",axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oe=OneHotEncoder(sparse_output=False)
```

```python
cat_cols=["state","district","market","commodity","variety","grade"]
```

```python
encoded_cols=oe.fit_transform(df[cat_cols])
```

```python
encoded_cols
```

```python
encoded_cols.shape
```

```python
type(encoded_cols)
```

```python
encoded_cols=pd.DataFrame(encoded_cols,columns=oe.get_feature_names_out(cat_cols))
```

```python
encoded_cols
```

```python
new_df=pd.concat([df,encoded_cols],axis=1)
```

```python
new_df.drop(cat_cols,axis=1,inplace=True)
```

```python
percentile25=df["min_price"].quantile(0.25)
percentile75=df["min_price"].quantile(0.75)
```

```python
print(percentile25)
print(percentile75)
```

```python
iqr=percentile75-percentile25
```

```python
iqr
```

```python
upper_limit=percentile75+(1.5*iqr)
lower_limit=percentile25-(1.5*iqr)
```

```python
upper_limit
```

```python
lower_limit
```

```python
new_df['min_price']=np.where(new_df['min_price']>upper_limit,
                    upper_limit,
                    np.where(new_df['min_price']<lower_limit,
                             lower_limit,
                             new_df['min_price']
                             )
                    )
```

```python
sns.boxplot(new_df["min_price"])
```
