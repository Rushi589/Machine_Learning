**OHE using sklearn's ohe class**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
df_to_encode=['gender','ssc_board','hsc_board','hsc_subject','undergrad_degree','work_experience','specialisation']
```

```python
encoded_df=ohe.fit_transform(df[df_to_encode]).toarray()
```

```python
encoded_df
```

```python
encoded_df.shape
```

```python
encoded_df=pd.DataFrame(encoded_df,columns=ohe.get_feature_names_out(df_to_encode))
```

```python
type(encoded_df)
```

```python
df=pd.concat([df,encoded_df],axis=1)
```

```python
df
```

```python
df.drop(df_to_encode,axis=1,inplace=True)
```

```python
df
```

**Ordinal Encoding**

```python
odf=pd.DataFrame({'review':['average','poor','good','good','average'],'education':['school','ug','pg','pg','ug']})
```

```python
odf
```

```python
from sklearn.preprocessing import OrdinalEncoder
order_encode=OrdinalEncoder(categories=[['poor','average','good'],['school','ug','pg']])
```

```python
odf=order_encode.fit_transform(odf)
```

```python
odf=pd.DataFrame(odf,columns=order_encode.get_feature_names_out())
```
